```python
def sigmoid(H):
    return 1/(1+math.pow(math.e,-H))
```

```python
# generate random data-set
np.random.seed(0)
noise = np.random.rand(100, 1)
rng = np.random.RandomState(1)
X = rng.rand(100, 2)
Y = 0.5 + np.dot(X, [1.5, -1.])
```

```python
import numpy as np
import  math

layer1=np.zeros(2) # H
layer2=np.zeros(1)
layer1_weight = [0.5, 0.5, 0.5 ,0.5]
layer2_weight = [0.4, 0.45]
bias=[0.35, 0.6]
learnRate=0.5

for echo in range(1):
    eTotal=0
    for n in range(6):
        x=X[n]
        y=[Y[n]]
        # Forward Pass
        ## First layer
        for i in range(2):
            for j in range(2):
                layer1[i] += (x[j]*layer1_weight[(i*2)+j])

        ### calc first layer active function
        for i in range(2):
            layer1[i]=sigmoid(layer1[i]+bias[0])

        ## Second layer
        for i in range(1):
            for j in range(2):
                layer2[i] += (layer1[j]*layer2_weight[(i*2)+j])
        ### calc second layer active function
        for i in range(1):
            layer2[i]=sigmoid(layer2[i]+bias[1])

        ## calculating error
        eTotal+=(0.5*math.pow(layer2[0]-y[0],2))

    # #     print(layer2)
    # #     print(layer1)

        partial_layer2 = [layer2[0]*(1-layer2[0])*(y[0]-layer2[0])]
        partial_layer1 = [layer1[0]*(1-layer1[0])*(partial_layer2[0]*layer2_weight[0]), layer1[1]*(1-layer1[1])*(partial_layer2[0]*layer2_weight[1])]

        # Bbackward
        ## Second layer
        for i in range(1):
            for j in range(2):
                layer2_weight[(i*2)+j]+=learnRate*partial_layer2[i]*layer1[j]
        ## First layer
        for i in range(2):
            for j in range(2):
                layer1_weight[(i*2)+j]+=learnRate*partial_layer1[i]*x[j]
        if(n==0):
            print(layer2)
            print(y)
        

    print(eTotal/100)
#     print(layer2)
#     print(y)
    print("---------")
```
