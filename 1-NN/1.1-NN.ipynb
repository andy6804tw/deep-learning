```python
import numpy as np
import  math

# x1=np.array([0.05])
# x2=np.array([0.1])
x=[0.05,0.1]
# y1=np.array([0.01])
# y2=np.array([0.99])
y=[0.01, 0.99]

layer1=np.zeros(2) # H
layer2=np.zeros(2)
layer1_weight = [0.15, 0.2, 0.25, 0.3]
layer2_weight = [0.4, 0.45, 0.5, 0.55]
bias=[0.35, 0.6]
learnRate=0.5
```

```python
def sigmoid(H):
    return 1/(1+math.pow(math.e,-H))
```

```python
# Forward Pass
## First layer
for i in range(2):
    for j in range(2):
        layer1[i] += (x[j]*layer1_weight[(i*2)+j])
        
### calc first layer active function
for i in range(2):
    layer1[i]=sigmoid(layer1[i]+bias[0])
    
## Second layer
for i in range(2):
    for j in range(2):
        layer2[i] += (layer1[j]*layer2_weight[(i*2)+j])
### calc second layer active function
for i in range(2):
    layer2[i]=sigmoid(layer2[i]+bias[1])
    
## calculating error
eTotal=(0.5*math.pow(layer2[0]-y[0],2))+(0.5*math.pow(layer2[1]-y[1],2))
```

```python
# Bbackward
## Second layer
for i in range(2):
    for j in range(2):
        layer2_weight[(i*2)+j]+=learnRate*partial_layer2[i]*layer1[j]
## First layer
for i in range(2):
    for j in range(2):
        layer1_weight[(i*2)+j]+=learnRate*partial_layer1[i]*x[j]
```

```python
import numpy as np
import  math

# x1=np.array([0.05])
# x2=np.array([0.1])
x=[0.05,0.1]
# y1=np.array([0.01])
# y2=np.array([0.99])
y=[0.01, 0.99]

layer1=np.zeros(2) # H
layer2=np.zeros(2)
layer1_weight = [0.15, 0.2, 0.25, 0.3]
layer2_weight = [0.4, 0.45, 0.5, 0.55]
bias=[0.35, 0.6]
learnRate=0.5

for echo in range(10000):
    # Forward Pass
    ## First layer
    for i in range(2):
        for j in range(2):
            layer1[i] += (x[j]*layer1_weight[(i*2)+j])

    ### calc first layer active function
    for i in range(2):
        layer1[i]=sigmoid(layer1[i]+bias[0])

    ## Second layer
    for i in range(2):
        for j in range(2):
            layer2[i] += (layer1[j]*layer2_weight[(i*2)+j])
    ### calc second layer active function
    for i in range(2):
        layer2[i]=sigmoid(layer2[i]+bias[1])

    ## calculating error
    eTotal=(0.5*math.pow(layer2[0]-y[0],2))+(0.5*math.pow(layer2[1]-y[1],2))

    partial_layer2 = [layer2[0]*(1-layer2[0])*(y[0]-layer2[0]), layer2[1]*(1-layer2[1])*(y[1]-layer2[1])]
    partial_layer1 = [layer1[0]*(1-layer1[0])*((partial_layer2[0]*layer2_weight[0])+(partial_layer2[1]*layer2_weight[2])), layer1[1]*(1-layer1[1])*((partial_layer2[0]*layer2_weight[1])+(partial_layer2[1]*layer2_weight[3]))]

    # Bbackward
    ## Second layer
    for i in range(2):
        for j in range(2):
            layer2_weight[(i*2)+j]+=learnRate*partial_layer2[i]*layer1[j]
    ## First layer
    for i in range(2):
        for j in range(2):
            layer1_weight[(i*2)+j]+=learnRate*partial_layer1[i]*x[j]

    print(eTotal)
    print(layer2)
    print(y)
    print("---------")
```
